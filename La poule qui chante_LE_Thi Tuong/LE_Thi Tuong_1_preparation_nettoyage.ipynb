## 2. Nettoyage des données

```python
dispo_ali = pd.read_csv("DisponibiliteAlimentaire_2017.csv")
pop = pd.read_csv ("Population_2000_2018.csv")
```

```python
pop.head()
```

```python
pop.isna().sum()
```

```python
pop[pop["Zone"]=="China"]
```

```python
# Affiche les data de pop avec sélectionné les variables nécessaires

pop = pop[['Zone', 'Année', 'Valeur']]
pop1 = pop.pivot_table(index='Zone', columns='Année', values='Valeur')

pop1.head()

#pop1.rename({"Pays":"Zone", 2017 : "pop2017"}, axis=1, inplace=True)
```

```python
pop1["%pop change 2006-2017"]=((pop1[2017]-pop1[2006])/pop1[2006]*100)
pop1.head()
```

```python
pop1.rename({2017 : "pop2017"}, axis=1, inplace=True)
```

```python
pop1.head()
```

```python
pop_final = pop1[['pop2017', '%pop change 2006-2017']]
pop_final.head() 
```

```python
# Affiche les data d'alimentaire
dispo_ali_vol = pd.read_csv('AlimentaireVolailles_2017.csv')
```

```python
dispo_ali_vol
```

```python
dispo_ali
```

```python
dispo_ali_vol = dispo_ali_vol.pivot_table(index=['Zone', 'Produit'],
                                   columns='Élément', values='Valeur')
```

```python
dispo_ali_vol
```

```python
# Changer le nom pour les colonnes
dispo_ali_vol.rename(columns={"Disponibilité alimentaire (Kcal/personne/jour)":"DispoAliVolKcal",
                               "Disponibilité de protéines en quantité (g/personne/jour)":"DispoAliProtVol"},
                     inplace=True)
```

```python
dispo_ali_vol
```

```python
dispo_ali_vol = dispo_ali_vol[['DispoAliVolKcal','DispoAliProtVol']]
```

```python
dispo_ali_vol
```

```python
#dispo_ali = dispo_ali.pivot_table(index=['Zone', 'Produit'],
                                   #columns=['Élément'], values=['Valeur'])
dispo = dispo_ali.loc[dispo_ali['Produit']=='Viande de Volailles',:]
dispo['Valeur(kg)'] = dispo['Valeur']*1000
dispo.drop(['Produit', 'Unité','Valeur'], axis=1, inplace=True)

elements = ['Disponibilité alimentaire (Kcal/personne/jour)', 'Disponibilité de protéines en quantité (g/personne', 'Production', 'Importations - Quantité', 'Exportations - Quantité']
dispo_final = dispo.loc[dispo['Élément'].isin(elements)]
dispo_final
```

```python
dispo_final=dispo_final [['Zone', 'Élément', 'Valeur(kg)']]
```

```python
dispo_final.head()
```

```python
# Merges deux data
dispo_ali1 = pd.merge(dispo_ali_vol,dispo_final, on='Zone', how='inner')
dispo_ali1.head()
```

```python
# Merge: disponibilités et population
dispo_ali2 = pd.merge(dispo_ali1, pop_final, on='Zone', how='outer')
```

```python
dispo_ali2
```

```python
dispo_ali2.info
```

```python
#Calcul du ratio de la disponibilité en protéines d'origine vollaie sur la disponibilité totale en protéines
dispo_ali2['RatioProtVol%'] = np.round(dispo_ali2['DispoAliProtVol'] / dispo_ali2['DispoAliVolKcal']*100,2)

dispo_ali2.head()
```

```python
dispo_ali2['Élément'].unique()
```

```python
dispo_ali2 = dispo_ali2.loc[dispo_ali2['Élément']=='Disponibilité alimentaire (Kcal/personne/jour)',:]
```

```python
dispo_ali2.head()
```

```python
len(dispo_ali2)
```

```python
#Suppression des colonnes non utilisés
dispo_ali2 = dispo_ali2[['Zone','pop2017', '%pop change 2006-2017', 'RatioProtVol%',
                           'DispoAliVolKcal', 'DispoAliProtVol']]

dispo_ali2.head()
```

```python
pib_stab = pd.read_excel('pib_stab.xlsx')
pib_stab.head()
```

```python
stab = pib_stab[['Zone','Produit','Valeur']]
```

```python
stab
```

```python
stab1= pd.pivot_table(stab, index=['Zone'], columns=['Produit'],values=['Valeur'])
```

```python
stab1
```

```python
# Jointure avec dispo_ali2
dispo_ali3 = pd.merge(dispo_ali2, stab1, on='Zone', how='inner')
```

```python
dispo_ali3
```

```python
# Changer le nom pour les colonnes
dispo_ali3.rename(columns={"(Valeur, PIB par habitant)":"PIB/hab",
                               "(Valeur, Stabilité politique)":"stab_poli"},
                     inplace=True)
```

```python
dispo_ali3
```

```python
distance = pd.read_excel('distance.xlsx')
distance.head()
```

```python
len(distance)
```

```python
# Jointure avec la distance
dispo_stab_pop_dist = pd.merge(dispo_ali3,distance, on = 'Zone', how= 'left')
```

```python
dispo_stab_pop_dist.head()
```

```python
len(dispo_stab_pop_dist)
```

```python
dispo_stab_pop_dist.dropna()
```

```python
# affichage du nombre de valeurs manquantes
dispo_stab_pop_dist.isna().sum()
```

```python
dispo_stab_pop_dist.to_csv('df_final.csv')
```
